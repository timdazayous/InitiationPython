### Pierre-feuille-ciseaux

* Jeu simple contre l'ordinateur.
* Utilise des variables, des entrées utilisateur, la fonction random.choice pour le choix de l'ordinateur, et des instructions conditionnelles pour déterminer le gagnant.


```python
import random 

def Pierre_Feuille_Ciseaux ():
    # Declaration des variables

    liste_de_coups = [(1,"Pierre"), (2,"Feuille"), (3,"Ciseaux")]
    #dico_des_coups = {
    #    1:"Pierre",
    #    2:"Feuille",
    #    3:"Ciseaux"
    #}
    dico_des_coups = dict(liste_de_coups)

    partie_terminee = False
    #winner = "Bot"
    #coup_ordinateur = random.choice(1,2,3)

    print("Nouvelle partie de pierre feuille ciseaux !")

    while not partie_terminee:
        print("Choississez votre coup :")
        print("Pierre tapez: 1")
        print("Feuille tapez: 2")
        print("Ciseaux tapez: 3")

        choix_joueur = 0
        choix_joueur = input("Votre choix :")
        
        if choix_joueur.isdigit() and 1 <= int(choix_joueur) <= 3:
            choix_joueur = int(choix_joueur)

            coup_joueur = dico_des_coups[choix_joueur]

            coup_ordinateur = random.choice(dico_des_coups)
            partie_terminee = True

            #cle_joueur, valeur_joueur = coup_joueur 
            #cle_bot, valeur_bot = coup_ordinateur
            match (coup_joueur, coup_ordinateur):
                # Egalite
                case (x,y) if x == y:
                    print("Vous avez joué :", coup_joueur)
                    print("L'ordinateur a joué :", coup_ordinateur)
                    print("Egalité !!!")
                case ("Pierre", "Feuille"):
                    print("Vous avez joué :", coup_joueur)
                    print("L'ordinateur a joué :", coup_ordinateur)
                    print("Bravo vous avez gagné !!! :)")
                case ("Feuille", "Pierre"):
                    print("Vous avez joué :", coup_joueur)
                    print("L'ordinateur a joué :", coup_ordinateur)
                    print("Bravo vous avez gagné !!! :)")
                case ("Ciseaux", "Feuille"):
                    print("Vous avez joué :", coup_joueur)
                    print("L'ordinateur a joué :", coup_ordinateur)
                    print("Bravo vous avez gagné !!! :)")
                case _:
                    print("Vous avez joué :", coup_joueur)
                    print("L'ordinateur a joué :", coup_ordinateur)
                    print("Dommage vous avez perdu... :'(")
        else:
            print("Saississez un chiffre !")
        

Pierre_Feuille_Ciseaux()
```
