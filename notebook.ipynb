````python
import os

PATH = r'.venv'
Nfiles = 0
Ndirs = 0


for root, dirnames, filenames in os.walk(PATH):
    Ndirs = Ndirs + len(dirnames)
    Nfiles += len(filenames)

print("Ndirs", Ndirs)
print("Nfiles", Nfiles)
````

````python
from os import walk as oswalk

PATH = r'.venv'
Nfiles = 0
Ndirs = 0


for root, dirnames, filenames in oswalk(PATH):
    Ndirs = Ndirs + len(dirnames)
    Nfiles += len(filenames)

print("Ndirs", Ndirs)
print("Nfiles", Nfiles)
````

````python
from matplotlib import pyplot as plt
import numpy as np

debut = 0
fin = 10
step = 1
i = range(debut, fin, step)

for n in i:
    print(n)

debut = 0
fin = 100
# Par defaut 50
nbpoint = 10
# Pour les nombres a virgules egalement
t = np.linspace(debut, fin, nbpoint)
````

````python
fig = plt.figure(figsize=(2,2))

a = 1
b = 0
c = 1.2
d = -.1

funct = a*t + b
funct2 = c*t + d

plt.plot(t, funct, "-.", color = "red", lw = 3, label = "f1")
plt.plot(t, funct2, ":", color="blue", lw=3, label="f2")
plt.grid()
plt.xlim((t.min(), t.max()))
plt.xlim((np.minimum(funct.min(), funct2.min()), np.maximum(funct.max(), funct2.max())))
plt.legend()
````
