### Mini-jeu de devinettes
* L'ordinateur choisit un nombre aléatoire entre 1 et 100.
* L'utilisateur doit deviner le nombre en un nombre limité de tentatives.
* L'ordinateur donne des indices (plus grands, plus petits) à chaque tentative.


```python
import random

def Devinette_Mini_Jeu ():
    nomrbre_a_trouver = random.randint(0,100)
    partie_en_cours = True
    nombre_propose = 0
    nombre_tentatives_max = 5
    nombre_tentatives = 1

    print("Tentez de deviner le nombre choisit par le bot en !")
    while partie_en_cours and nombre_tentatives <= nombre_tentatives_max:
        nombre_propose = input("Saississez un nombre entre 0 et 100 compris :")

        if nombre_propose.isdigit() and 1 <= int(nombre_propose) <= 100:
            nombre_propose = int(nombre_propose)

            if nomrbre_a_trouver < nombre_propose:
                nombre_tentatives += 1
                print("Le nombre à deviner est plus petit que :", nombre_propose)
                continue
            elif nomrbre_a_trouver > nombre_propose:
                nombre_tentatives += 1
                print("Le nombre à deviner est plus grand que :", nombre_propose)
                continue
            else:
                nombre_tentatives += 1
                print("Bravo vous avez deviné le nombre :", nomrbre_a_trouver)

                partie_en_cours = False

    if nombre_tentatives > nombre_tentatives_max:
        print("Dommage vous avez atteint le nombre max d'essais !")
        print("Le nombre à deviner etait :", nomrbre_a_trouver)

Devinette_Mini_Jeu()
```
